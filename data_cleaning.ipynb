```python
import pandas as pd
```

```python
file_path = 'data/survey.xlsx'  # Replace with your actual file path
data = pd.read_excel(file_path)
data.head()
```

```python
normalized_data = data.copy()
```

```python
# Mapping categorical responses to numerical values
frequency_mapping_1 = {
    'Multiple times a day': 5,
    'Once a day': 4,
    'A few times a week': 3,
    'Once a week': 2,
    'Rarely': 1
}

frequency_mapping_2 = {
    'Always': 4,
    'Most of the time': 3,
    'Sometimes': 2,
    'Rarely': 1
}

frequency_mapping_3 = {
    'Always': 4,
    'Often': 3,
    'Sometimes': 2,
    'Rarely': 1
}

time_spent_mapping = {
    'More than 3 hours': 5,
    '2-3 hours': 4,
    '1-2 hours': 3,
    '30 minutes to 1 hour': 2,
    'Less than 30 minutes': 1
}

engagement_mapping = {
    '5. Fully engaged (Went beyond expectations, actively participated in all activities, consistently completed tasks ahead of deadlines, took initiative beyond assigned tasks)': 5,
    '4. Highly engaged (Fulfilled all the expectations)': 4,
    '3. Moderately engaged (Participated in most activities, completed most tasks on time, put in adequate effort but still can be improved)': 3,
    '2. Minimally engaged (Participated in some activities and completed some assigned tasks on time)': 2,
    '1. Not engaged at all (Rarely participated in activities or completed assigned tasks on time)': 1
}

binary_mapping = {'Yes': 1, 'No': 0}
```

```python
normalized_data['Login Frequency'] = normalized_data['How often do you log into Canvas this summer?'].map(frequency_mapping_1)
normalized_data['Time Spent on Canvas'] = normalized_data['How much time do you spend on Canvas weekly for internship-related activities?  '].map(time_spent_mapping)
normalized_data['Accessed Optional Module'] = normalized_data['Have you accessed the Optional Continuing Education module in the Canvas Open Learning page before?'].map(binary_mapping)
normalized_data['Check for Emails'] = normalized_data['How often do you check for emails and announcements?'].map(frequency_mapping_1)
normalized_data['Timely Submission'] = normalized_data['How often do you submit the Canvas activities on time?'].map(frequency_mapping_2)
normalized_data['Time on Client Project'] = normalized_data['How much time do you spend weekly on your client project?  '].map(time_spent_mapping)
normalized_data['Time on SMART Goals'] = normalized_data['How much time do you spend weekly working on the three SMART goals set at the beginning of the internship?  '].map(time_spent_mapping)
normalized_data['Communication with Coordinator'] = normalized_data['How often do you communicate with your ATLAS coordinator?'].map(frequency_mapping_1)
normalized_data['Communication with Client'] = normalized_data['How often do you communicate with your client?'].map(frequency_mapping_1)
normalized_data['Communication with Interns'] = normalized_data['How often do you communicate and collaborate with your coworkers or other interns on projects?'].map(frequency_mapping_1)
normalized_data['Internship Meaningful'] = normalized_data['How often do you feel that your internship tasks are meaningful and impactful?'].map(frequency_mapping_3)
normalized_data['Internship Difficulty'] = normalized_data['How often do you feel challenged by your internship tasks?  '].map(frequency_mapping_3)

normalized_data['Learned From Internship'] = normalized_data['On a scale of 1 to 5,  how much have you learned from your internship so far? '].astype(int)
normalized_data['Satisfaction With Support'] = normalized_data['On a scale of 1 to 5, how satisfied are you with the level of support and resources provided during your internship? '].astype(int)
normalized_data['Overall Engagement'] = normalized_data['How would you rate your overall level of engagement during the internship program?'].map(engagement_mapping)
```

```python
normalized_data = normalized_data[['Login Frequency', 'Time Spent on Canvas', 'Accessed Optional Module', 'Check for Emails', 
                                   'Timely Submission', 'Time on Client Project', 'Time on SMART Goals', 'Communication with Coordinator', 
                                   'Communication with Client', 'Communication with Interns', 'Internship Meaningful', 'Internship Difficulty', 
                                   'Learned From Internship', 'Satisfaction With Support', 'Overall Engagement']]
```

```python
normalized_data.to_excel('data/normalized_survey_data.xlsx', index=False)
```
